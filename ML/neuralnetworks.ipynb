```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


```

```python
# Read the dataset
data = pd.read_csv('Churn_Modelling.csv')

data
```

```python
data = data.drop(['RowNumber', 'Surname', 'CustomerId'], axis= 1) #Dropping the unnecessary columns
```

```python
X = data[['CreditScore','Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']]
states = pd.get_dummies(data['Geography'],drop_first = True)
gender = pd.get_dummies(data['Gender'],drop_first = True)
```

```python
data = pd.concat([data,gender,states], axis = 1)
```

```python
# Separate the features and target variable
X = data[['CreditScore','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary','Male','Germany','Spain']]
y = data['Exited']
```

```python
# Divide the data into training and test sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30)

```

```python
sc = StandardScaler()
```

```python
# Normalize the train and test data

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

```python
# Initialize and build the model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

```

```python
# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32)

```

```python
# Evaluate the model
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)

```

```python
# Print the accuracy score and confusion matrix
from sklearn.metrics import accuracy_score, confusion_matrix

```

```python
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)

confusion_matrix = confusion_matrix(y_test, y_pred)
print('Confusion matrix:')
print(confusion_matrix)
```
